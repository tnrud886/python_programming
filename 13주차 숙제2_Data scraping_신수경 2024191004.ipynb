```python
top10_books = books.head(10)

def get_isbn(row):
    isbn = row['isbn13']
    return get_page_cnt(isbn)

page_count = top10_books.apply(get_isbn, axis=1)
print(page_count)
```

```python
top10_with_page_count = pd.merge(top10_books, page_count, left_index=True, right_index=True)
print(top10_with_page_count)
```

```python
def get_page_cnt(isbn):
    url = f'https://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')

    # 상세 페이지 링크 추출
    prd_link = soup.find('a', attrs={'class': 'gd_name'})
    url = 'https://www.yes24.com' + prd_link['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')

    # 제목 추출
    title = soup.find('h2', attrs={'class': 'gd_name'}).text.strip()

    # 쪽수 추출
    prod_detail = soup.find('div', attrs={'id': 'infoset_specific'})
    prod_tr_list = prod_detail.find_all('tr')
    page_td = ''
    for tr in prod_tr_list:
        if tr.find('th').text == '쪽수, 무게, 크기':
            page_td = tr.find('td').text
            break

    page = page_td.split()[0] if page_td else '정보 없음'
    return page, title
```

```python
page, title = get_page_cnt('9788934985068')
print(f"제목: {title}, 쪽수: {page}")
```
